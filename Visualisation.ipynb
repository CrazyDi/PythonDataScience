# Визуализация данных

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
student_performance = pd.read_csv('data/StudentsPerformance.csv')
student_performance = student_performance.rename(columns = {
    'parental level of education': 'parental_level_of_education',
    'test preparation course': 'test_preparation_course', 
    'math score': 'math_score', 
    'reading score': 'reading_score',
    'writing score': 'writing_score'
})
student_performance.head()
```

## Встроенная визуализация pandas

Простейшая гистограмма

```python
student_performance.math_score.hist()
```

Нарисуем скаттерплот

```python
student_performance.plot.scatter(x='math_score', y='reading_score')
```

## Библиотека seaborn

Построим тот же скаттерплот

```python
sns.lmplot(x='math_score', y='reading_score', data=student_performance)
```

Добавим выделение групп цветом

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=student_performance)
```

Уберем регрессионную прямую

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=student_performance, fit_reg=False)
```

Подпишем оси

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=student_performance, fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
